```python
import pyproj
```

```python
lambert = pyproj.Proj("+init=EPSG:31370")
wgs84 = pyproj.Proj("+init=EPSG:4326")
```

```python
pyproj.transform(lambert, wgs84,221651,44263)
```

```python
def transform_lambert_to_wgs(row):
    """
    Converts the x and y coordinates of this row into a Series of the
    longitude and latitude.
    
    """
    return pd.Series(pyproj.transform(lambert, wgs84, row['X'], row['Y']))
```

```python
transform_lambert_to_wgs(survey_data.loc[0])
```

```python
survey_data.apply(transform_lambert_to_wgs, axis=1)
```

```python
survey_data[["decimalLongitude" ,"decimalLatitude"]] = survey_data.apply(transform_lambert_to_wgs, axis=1)
```

```python
def species_on_map(df, background='Stamen Terrain'):
    """plot species occurrences on a map with markers
    
    Parameters
    ----------
    df: Pandas.DataFrame
        dataframe with DarwinCore terms. For coorindates, the column names decimalLatitude and decimalLongitude will be used.
    background: str
        valid name for folium background tiles
    """
    map_1 = fol.Map(location=df[['decimalLatitude', 'decimalLongitude']].mean().values.tolist(),
                   zoom_start=8,
                   tiles=background)
    for ids, row in df.iterrows():
        fol.Marker(row[['decimalLatitude', 'decimalLongitude']], popup=row["scientificName"]).add_to(map_1)
    return map_1
```

```python
species_on_map(survey_data, background='Stamen Terrain')
```

```python
import folium as fl
from folium.plugins import MarkerCluster
from folium.map import Marker
```

```python
make_a_map = fl.Map(tiles='OpenStreetMap')
lons = survey_data["decimalLongitude"].values
lats = survey_data["decimalLatitude"].values

locations = np.vstack((lats, lons)).transpose()  #list(zip(lats, lons))
locations = locations.reshape(int(locations.size/2), 2)
popups = survey_data['scientificName']

make_a_map.add_children(MarkerCluster(locations=locations, popups=popups))
make_a_map.fit_bounds([(min(lats), min(lons)), (max(lats), max(lons))])
```
